```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import pandas as pd
from surprise import Dataset
from surprise import Reader
from surprise import Dataset, evaluate
from surprise import SVD
from surprise.model_selection import KFold
from surprise import accuracy
```

```python
#Reading data through pandas dataframe
review_text_dataframe = pd.read_csv('~/AnacondaProjects/Project_256/Hotel_Reviews.csv')
review_text_dataframe
```

```python
review_text_dataframe1 = review_text_dataframe.drop(['Review_Date','Review_Total_Negative_Word_Counts','Reviewer_Nationality','Negative_Review','Positive_Review','Tags','days_since_review','lat','lng'], axis=1)

review_text_dataframe1
```

```python
#Creating the suprise dataset from pandas dataframe and tranining the dataset
reader = Reader(rating_scale=(0, 110))
data = Dataset.load_from_df(review_text_dataframe1[['Hotel_Name', 'Average_Score','Reviewer_Score']], reader)
trainingSet = data.build_full_trainset()
```

```python
print(data)
```

```python
#Using SVD algo and calculating Test Time , RMSE, MAE
from surprise.model_selection import cross_validate
algo1=SVD(n_epochs=50,lr_all=0.001,  reg_all=0.008)

kf = KFold(n_splits=3)
print(cross_validate(algo1, data, measures=['RMSE', 'MAE'], cv=5, verbose=True))

for trainset, testset in kf.split(data):

    # train and test algorithm.
    algo1.fit(trainingSet)
    predictions = algo1.test(testset)
    accuracy.rmse(predictions, verbose=True)
```

```python
#Using KNNBasic algo and calculating Test Time , RMSE, MAE
from surprise import KNNBasic
sim_options = {'name': 'cosine',
               'user_based': False  # compute  similarities between items
               }
algo = KNNBasic(sim_options=sim_options)

kf = KFold(n_splits=5)
print(cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True))
for trainset, testset in kf.split(data):

    # train and test algorithm.
    algo1.fit(trainingSet)
    predictions = algo1.test(testset)
    accuracy.rmse(predictions, verbose=True)
```

```python
#Using CoClusterings algo and calculating Test Time , RMSE, MAE
from surprise.prediction_algorithms import CoClustering

algo = CoClustering()

kf = KFold(n_splits=5)
print(cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True))
for trainset, testset in kf.split(data):

    # train and test algorithm.
    algo1.fit(trainingSet)
    predictions = algo1.test(testset)
    accuracy.rmse(predictions, verbose=True)
```
