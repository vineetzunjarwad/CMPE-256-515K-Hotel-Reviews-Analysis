```python
#Open Source Data Analysis Library
import pandas as pd
```

```python
#Reading data through pandas dataframe
review_text_dataframe = pd.read_csv('~/AnacondaProjects/Project_256/Hotel_Reviews.csv')
review_text_dataframe
```

```python
review_text_dataframe1 = review_text_dataframe.groupby(['Reviewer_Nationality']).agg({
                             'Negative_Review': ', '.join,'Positive_Review': ', '.join}).reset_index()
review_text_dataframe1
```

```python
#creating new column review text which is combination of negative and positive coloumn
review_text_dataframe1['review_text'] = review_text_dataframe1['Positive_Review'].astype(str) + review_text_dataframe1['Negative_Review'].astype(str)
```

```python
review_text_dataframe1
```

```python
#method to find similarity between reviews of two hotels
from sklearn.feature_extraction.text import TfidfVectorizer

#Removing the stop words and combining all the CountVectorizer and TfidfTransformer
tf_idf_vectorizer = TfidfVectorizer(stop_words='english')

#Replace Na words with '' 
review_text_dataframe1['review_text'] = review_text_dataframe1['review_text'].fillna('')

#Learn vocabulary and idf  and return term-document matrix
term_document_matrix = tf_idf_vectorizer.fit_transform(review_text_dataframe1['review_text'])

#Print the shape of tfidf_matrix which has around 8100 words
term_document_matrix.shape
```

```python
#used cosine similarity method to compute the similarity between hotels
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(term_document_matrix, term_document_matrix)
```

```python
#creating a Series with index = Reviewer_Nationality
import numpy as np
review_text_dataframe1['Reviewer_Nationality']=review_text_dataframe1['Reviewer_Nationality'].str.strip()
list_Nationality = pd.Series(review_text_dataframe1.index, index=review_text_dataframe1['Reviewer_Nationality']).drop_duplicates()
list_Nationality['Australia']
```

```python
# Method which takes the index of user hotel as input and calculate the similar hotels
def recommend(idx,cosine_sim=cosine_sim):

    # Calculate the similarrity of all hotel with given hotel
    similarity = list(enumerate(cosine_sim[idx]))

    #Sort the similarity
    similarity_list = sorted(similarity, key=lambda x: x[1], reverse=True)

    # Store the similarity score in list
    similarity_list = similarity_list[1:11]

    # Get the hotel indices
    hotels_similar = [i[0] for i in similarity_list]

    # Return the top 10 most similar hotel
    return review_text_dataframe1['Reviewer_Nationality'].iloc[hotels_similar]
```

```python

#user input hotel name 11 cadogan garden and get top 10 similar hotel based on user review
country='Albania'
for index_val, series_val in list_Nationality.iteritems():
    if country in index_val:     
        id=series_val
        print(id)
recommend(id)
```
