```python
# leisure_hotel
topCoupleHotels=dataset[dataset['Tags'].str.contains('Couple')].groupby('Hotel_Name')['Reviewer_Nationality'].count().sort_values(ascending=False).head(15)
topLeisureHotels = dataset[dataset['Tags'].str.contains('Leisure')].groupby('Hotel_Name')['Reviewer_Nationality'].count().sort_values(ascending=False).head(15)
result=pd.DataFrame({'Couple':pd.Series(topLeisureHotels),'Leisure':pd.Series(topCoupleHotels) })

result
```

```python
#Graph for Top Leisure Hotels
from matplotlib import pyplot as plt
plt.xlabel('Hotel_Name', fontweight='bold')
plt.ylabel('Leisure Score', fontweight='bold')
plt.title('Top Leisuire friendly Hotel', fontweight='bold', color='purple')
topLeisureHotels.head(15).plot.bar(x='Hotel Name',y='Ratings',figsize=(20,15),legend=False , color=[plt.cm.Paired(np.arange(len(topLeisureHotels)))]).set_ylim(1700,3550)
plt.show()
```

```python
#Business Trip Hotel

topBusinessHotels = dataset[dataset['Tags'].str.contains('Business trip')].groupby('Hotel_Name')['Reviewer_Nationality'].count().sort_values(ascending=False).head(15)
from matplotlib import pyplot as plt
plt.xlabel('Hotel_Name', fontweight='bold')
plt.ylabel('Business Trip Score', fontweight='bold')
plt.title('Top Business Trip Hotel', fontweight='bold', color='purple')
topBusinessHotels.head(15).plot.bar(x='Hotel Name',y='Ratings',figsize=(20,15),legend=False , color=[plt.cm.Paired(np.arange(len(topLeisureHotels)))]).set_ylim(300,1500)
plt.show()
```
