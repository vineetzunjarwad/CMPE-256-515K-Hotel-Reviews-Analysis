```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
dataset = pd.read_csv('~/AnacondaProjects/Project_256/Hotel_Reviews.csv')
```

```python
dataset
```

```python
print(dataset['Reviewer_Score'].isnull().values.sum().sum())
print(dataset['Hotel_Name'].isnull().values.sum().sum())
```

```python
#prnting the datatypes of column of dataset
dataset.dtypes
```

```python
#Plotted the grah of top rated hotels in Europe
list_of_top_hotel= dataset.groupby('Hotel_Name')['Reviewer_Score'].sum()
count=dataset.groupby('Hotel_Name')['Reviewer_Score'].count()
result= list_of_top_hotel/count
result1=result.sort_values(ascending= False).head(15)
result1
```

```python
from matplotlib import pyplot as plt
plt.xlabel('Hotel_Name', fontweight='bold')
plt.ylabel('Ratings', fontweight='bold')
plt.title('Hotel Ratings', fontweight='bold', color='purple')
result1.head(15).plot.bar(x='Hotel Name',y='Ratings',legend=False , color=[plt.cm.Paired(np.arange(len(result1.head(15))))]).set_ylim(9.5,9.9)
plt.show()
```

```python
best_hotel=result.head(15).index.values
best_hotel
```

```python
#Plotted the grah of variation in rating of reviewer score of top 15 hotels
for hotel_name in best_hotel:
    hotel_datset = dataset[dataset.Hotel_Name == hotel_name]
    hotel_datset.plot('Review_Date', 'Reviewer_Score',figsize=(10,7),color='g')
    plt.xlabel('Review_Date', fontweight='bold')
    plt.ylabel('Rating Change', fontweight='bold')
    plt.title('Change of Review Score for :'+hotel_name, color='purple', fontweight='bold')
    plt.show()
    
```
