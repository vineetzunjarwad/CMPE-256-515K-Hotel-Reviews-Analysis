```python
import pandas as pd
```

```python
#Read the csv file
visualization = pd.read_csv('~/AnacondaProjects/Project_256/Hotel_Reviews.csv')
visualization
```

```python
#calculated the visitors from different nations to europe
result = ((visualization.Reviewer_Nationality.value_counts()/len(visualization.Reviewer_Nationality))*100).sort_values(ascending = False)
result
```

```python
#Plotted the graph of the Visitors
from matplotlib import pyplot as plt

plt.title('Visitors Ratio', fontweight='bold', color='purple')
result.head(10).plot.pie(y='percentage of people visiting',subplots=True, figsize=(20,10))

plt.show()
```
